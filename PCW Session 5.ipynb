Write a Python implementation of the merge pseudocode (both parts). Your Python function must take the same four input arguments as described in the $MERGE(A,p,q,r)$. **Note**: Please note that the pseudocde in Cormen et. al. works with the indexing convention starting at 1; in Python we will use the customary indexing convention starting at 0.

````python

def merge(A, p, q, r):
    n1 = q - p + 1 #devide array A into two subarrays n1 and n2
    n2 = r - q 
    L = [0] * (n1) #we create two arrays
    R = [0] * (n2) 
    for i in range(0 , n1): #we then copy the two subarrays into L and R and define them
        L[i] = A[p + i-1] 
    for j in range(0 , n2): 
        R[j] = A[q + j] 
    L[n1 + 1] = float("inf") #we set this infinite value for termination
    R[n2 + 2] = float("inf")
    i = 0 #set initial values for the key and i and j
    j = 0
    k = 1
    while L[i] != L[n1 + 1]: #Terminate when we reach the last card
        if L[i] <= R[j]: #comparing the top numbers of both arrays
            A[k] = L[i] #put the smaller one into array A
            i = i + 1 #increase the index to next number for comparison
        else: 
            while R[j] != R[n2+2]: 
                A[k] = R[j]
                j = j + 1 #to go to the next unsorted number
                k = k + 1
    return merge(A, p, q, r)
    raise NotImplementedError()
````
